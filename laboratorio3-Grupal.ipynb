```python
import pandas as pd

#estudiantes:

#Francisco Arias Sanabria
#Josefat Arguedas Gutiérrez
#Gerardo Salazar Vargas
#Angie Soza Hidalgo

students = [('Ana',23,'San Jose','A'),
            ('Esteban',22,'Heredia','B'),
            ('Juan',22,'Cartago','A'),
            ('Karla',22,'Limón','A'),
            ('Ernesto',21,'Heredia','B'),
            ]

df = pd.DataFrame(students, columns =['Nombre','Edad','Ciudad','Seccion'])

print(df)

def getIndexes(dfObj,value):

  listOfPos=[]
  result = dfObj.isin([value])
  seriesObj = result.any()
  columnNames = list(seriesObj[seriesObj == True].index)

  for col in columnNames:
    rows = list(result[col][result[col]==True].index)

    for row in rows:
      listOfPos.append((row,col))

  return listOfPos


listOfPositions = getIndexes(df,'A')
print('Index positions of 22 in Dataframe: ')

for i in range(len(listOfPositions)):
  print(listOfPositions[i])


#getIndexes(df,'Heredia')
#getIndexes(df,'Cartago')
getIndexes(df,22)

df2=pd.read_csv('/content/coffee_analysis.csv')
data=pd.DataFrame(df2)
print(data)

getIndexes(df2,'A.R.C.')
getIndexes(df2,'Costa Rica')

```
