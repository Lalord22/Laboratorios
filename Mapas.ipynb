```python
from folium.plugins import MiniMap
from branca.element import Figure

# Crear el mapa con una ubicaci√≥n central
my_map = folium.Map(location=[9.9965632, -84.1165315], zoom_start=8, tiles='stamenterrain')

# Agregar el primer punto al mapa
popuptext = '<b>UNA</b>'
folium.Marker(location=[9.9965632, -84.1165315], popup=popuptext, icon=folium.Icon(color="red", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[9.9965632, -84.1165315], color="blue", weight=6, fill_color="white", fill_opacity=0.5, tooltip="UNA").add_to(my_map)

# Agregar el segundo punto al mapa
popuptext = '<b>Guanacaste</b>'
folium.Marker(location=[10.6175015,-85.4560314], popup=popuptext, icon=folium.Icon(color="red", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[10.6175015,-85.4560314], color="blue", weight=6, fill_color="white", fill_opacity=0.5, tooltip="Guanacaste").add_to(my_map)

# Agregar el tercer punto al mapa
popuptext = '<b>Perez</b>'
folium.Marker(location=[9.377374,-83.7382731], popup=popuptext, icon=folium.Icon(color="red", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[9.377374,-83.7382731], color="blue", weight=6, fill_color="white", fill_opacity=0.5, tooltip="Perez").add_to(my_map)

# Agregar el Cuarto punto al mapa
popuptext = '<b>Benjamin</b>'
folium.Marker(location=[9.9697604,-84.1316207], popup=popuptext, icon=folium.Icon(color="red", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[9.9697604,-84.1316207], color="blue", weight=6, fill_color="white", fill_opacity=0.5, tooltip="Benjamin").add_to(my_map)

# Agregar el quinto punto al mapa
popuptext = '<b>Benjamin</b>'
folium.Marker(location=[10.0197603,-84.1995895], popup=popuptext, icon=folium.Icon(color="red", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[10.0197603,-84.1995895], color="blue", weight=6, fill_color="white", fill_opacity=0.5, tooltip="Benjamin").add_to(my_map)


# Agregar el minimapa
minimap = MiniMap(tile_layer='stamenterrain', position='bottomright')
my_map.add_child(minimap)

my_map.save('C:mapa.html')
my_map
```
