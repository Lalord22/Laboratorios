```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/School-Age-Digital-Connectivity - Primary.csv")

#crear una table de contingencia(cuenta las frecuencias de combinaciones de income group)
contingency_table = pd.crosstab(df['Income Group'],df['Region'])

"""
Crear el mada de calor usando seaborn

como se puede visualizar, la libreir a tras la funcion heatmap

annot: anotaciones en la celda
cmap: permite personalizar el mapa de colores

"""

#Crear el mapa de calor
plt.figure(figsize=[10,7])
sns.heatmap(contingency_table, annot=True, cmap='coolwarm')

#Mostrar el grafico
plt.title("Mapa de calor de Income Group vs Residence-Urban")
plt.show()
```
