```python

#libreria de graficos
import matplotlib.pyplot as plt

#Otra libreria de graficacion
import seaborn as sns

 #Apertura de archivo
import pandas as pd
datos=pd.read_csv('/content/casasboston.csv')

print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]

df = datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS":"RIO CHARLES",
    "RM":"N_HABITACIONES_MEDIO",
    "MEDV":"VALOR_MEDIANO",
    "LSTAT":"PCT_CLASE_BAJA"
})

print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist( color = "silver", ec="green")

plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN", alpha=0.5)
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
valor_por_ciudad.head(10).plot.barh(color="green")
plt.show()
```

```python
table = pd.crosstab(df["INDICE_CRIMEN"],df["VALOR_MEDIANO"])
print(table)
```
