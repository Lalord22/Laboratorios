```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv("/content/coffee_analysis.csv")

print(datos)
```

```python
valor_por_tostado=datos.groupby("roast")["100g_USD"].mean()
valor_por_tostado.head(10).plot.barh()
plt.title('Promedio de cafe por tipo tostado')
plt.ylabel('Roast')
plt.xlabel('Valor promedio (USD por 100g)')
plt.show()

```
