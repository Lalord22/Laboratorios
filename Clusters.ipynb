```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# Josafat Arguedas
#Francisco Sanabria
#Gerardo salazar

# Cargar los datos
data = pd.read_csv("/content/Basic_Stats.csv")

# Eliminar columnas no numéricas
data_numeric = data.select_dtypes(include=[np.number])

# Manejar los valores NaN (por ejemplo, reemplazándolos con la media de la columna)
data_numeric = data_numeric.fillna(data_numeric.mean())

# Ajustar el algoritmo de KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(data_numeric)
print(data.columns)
# Visualizar los clústeres
plt.scatter(data_numeric["Age"], data_numeric["Height (inches)"], c=kmeans.labels_)
plt.show()
```
